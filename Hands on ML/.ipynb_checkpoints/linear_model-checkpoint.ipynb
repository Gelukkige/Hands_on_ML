```python
import numpy as np
from mglearn.datasets import make_wave
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = make_wave(n_samples=1000)
```

```python
plt.scatter(data[0], data[1], marker='.')
plt.show()
```

```python
features = data[0]
labels = data[1]
print('shape of feature:', features.shape)
print('shape of label:', labels.shape)
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
# linear regression
class linear_regression:
    def __init__(self, lr=0.01, max_iter=50, batch_size=20):
        self.lr = lr
        self.max_iter = max_iter
        self.batch_size = batch_size
    
    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        num_features = X.shape[1]
        self.history_loss = []
        self.w = np.random.normal(0, 1, size=(num_features, 1))
        self.b = np.zeros(1)
        
        # 小批量生成
        def data_iter(features, labels, batch_size):
            num_samples = features.shape[0]
            indices = list(range(num_samples))
            np.random.shuffle(indices)
            for i in range(0, num_samples, batch_size):
                batch_indices = np.array(indices[i:min(i + batch_size, num_samples)])
                yield features[batch_indices], labels[batch_indices]

        # 均方误差
        def mean_squared_loss(y_hat, y):
            return np.mean((y_hat - y.reshape(y_hat.shape))**2)
        
        # 小批量梯度下降
        def sgd(X, y, w, b, lr, batch_size):
            y_hat = np.dot(X, self.w) + self.b
            loss = mean_squared_loss(y_hat, y)
            error = y_hat - y.reshape(y_hat.shape)
            dw = 2 * np.dot(X.T, error) / self.batch_size
            db = 2 * np.mean(error)
            w -= self.lr * dw
            b -= self.lr  * db
            return w, b, loss
        
        for i in range(self.max_iter):
            for x_, y_ in data_iter(X, y, self.batch_size):
                self.w, self.b, loss = sgd(x_, y_, self.w, self.b, self.lr, self.batch_size)
            if loss <= 0.2:
                break
            self.history_loss.append(np.mean(loss))
        return self
    
    def history(self):
        return self.history_loss
    
    def predict(self, X):
        return np.dot(X, self.w) + self.b
    
    def score(self, X, y):
    pred_ = self.predict(X)
    pred_ = pred_.ravel()
    y_mean = np.mean(y)
    score = 1 - (np.sum((y - pred_)**2) / np.sum((y - y_mean)**2))
    return score
    
    def param_(self):
        return {'w:': self.w, 'b:': self.b}
```

```python
# ridge regression
class Ridge:
    def __init__(self, lr=0.01, alpha=0.5, max_iter=50):
        self.lr = lr
        self.alpha = alpha
        self.max_iter = max_iter
    
    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        num_features = X.shape[1]
        num_data = X.shape[0]
        self.w = np.random.normal(0, 1, size=(num_features, 1))
        self.b = np.zeros(1)
        self.loss_history = []
        self.accuracy = []
        
        def mean_square_loss(y_hat, y, w, alpha):
            return np.mean((y_hat - y)**2 + alpha * w**2)
        
        def gd(X, y, y_hat, w, b, lr,  alpha):
#             y_hat = np.dot(X, w) + b  # 预测值
            error  = y_hat - y.reshape(y_hat.shape)  # 误差
            dw = 2 * (np.dot(X.T, error) + alpha * w) / num_data
            db = 2 * np.mean(error)
            w -= lr * dw
            b -= lr * db
            return w, b
        
        for i in range(self.max_iter):
            y_hat = np.dot(X, self.w) + self.b  # 预测值
            loss = mean_square_loss(y_hat, y, self.w, self.alpha)  # 均方损失
            self.w, self.b = gd(X, y, y_hat, self.w, self.b, self.lr, self.alpha)
            self.loss_history.append(loss)
        return self
    
    def predict(self, X):
        return np.dot(X, self.w) + self.b
    
    def score(self, X, y):
        pred_ = self.predict(X)
        pred_ = pred_.ravel()
        y_mean = np.mean(y)
        score = 1 - (np.sum((y - pred_)**2) / np.sum((y - y_mean)**2))
        return score
    
    def history(self):
        return self.loss_history
    
    def param_(self):
        return ({'w: ': self.w, 'b: ': self.b})
```

```python
# lasso regression
class Lasso:
    def __init__(self, lr=0.01, alpha=0.5, max_iter=50):
        self.lr = lr
        self.alpha = alpha
        self.max_iter = max_iter
    
    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        num_features = X.shape[1]
        num_data = X.shape[0]
        self.w = np.random.normal(0, 1, size=(num_features, 1))
        self.b = np.zeros(1)
        self.loss_history = []
        self.accuracy = []
        
        def mean_square_loss(y_hat, y, w, alpha):
            return np.mean((y_hat - y)**2 + alpha * w**2)
        
        def gd(X, y, y_hat, w, b, lr,  alpha):
#             y_hat = np.dot(X, w) + b  # 预测值
            error  = y_hat - y.reshape(y_hat.shape)  # 误差
            dw = 2 * (np.dot(X.T, error) + alpha) / num_data
            db = 2 * np.mean(error)
            w -= lr * dw
            b -= lr * db
            return w, b
        
        for i in range(self.max_iter):
            y_hat = np.dot(X, self.w) + self.b  # 预测值
            loss = mean_square_loss(y_hat, y, self.w, self.alpha)  # 均方损失
            self.w, self.b = gd(X, y, y_hat, self.w, self.b, self.lr, self.alpha)
            self.loss_history.append(loss)
        return self
    
    def predict(self, X):
        return np.dot(X, self.w) + self.b
    
    def score(self, X, y):
        pred_ = self.predict(X)
        pred_ = pred_.ravel()
        y_mean = np.mean(y)
        score = 1 - (np.sum((y - pred_)**2) / np.sum((y - y_mean)**2))
        return score
    
    def history(self):
        return self.loss_history
    
    def param_(self):
        return ({'w: ': self.w, 'b: ': self.b})
```

```python
lasso = Lasso()
lasso.fit(x_train, y_train)
```

```python
pred = lasso.predict(x_test)
```

```python
plt.figure()
plt.plot(x_test, y_test, 'bo')
plt.plot(x_test, pred, 'ro')
plt.show()
```
